```python
ds['Review Rating']=ds.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
ds.isna().sum()
```

```python
ds.columns=ds.columns.str.lower()
ds.columns=ds.columns.str.replace(' ','_')
ds=ds.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
ds.isna().sum()
```

```python
#create column age group
labels=['Young Adult','Adult','Middle-aged','Senior']
ds['age_group']=pd.qcut(ds['age'],q=4,labels=labels)
```

```python
ds[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

ds['purchase_frequency_days'] = ds['frequency_of_purchases'].map(frequency_mapping)
```

```python
ds[['purchase_frequency_days','frequency_of_purchases']].head(10)
```
